```python
import random

print("00_guess_my_number")

def main():
    secret_number = random.randint(1, 100)
    print("I'm thinkiing of a number between 1 and 100.")

    guess = int(input("Enter a guess: "))

    while guess != secret_number:
        if guess < secret_number:
            print("Too low.")
        else:
            print("Too high.")

        guess = int(input("Enter a guess: "))  # Moved this outside the if-else block

    print(f"Congratulations! The number was {secret_number}")

if __name__ == "__main__":
    main()
```

```python
print("01_fibonacci")
max_value = 10000
def main():
  a,b = 0,1
  print(a,b,end=" ")
  while True:
    c= a+b
    if c>=max_value:
      break
    print(c,end=" ")
    a,b = b,c

if __name__ == "__main__":
  main()
```

```python
print("02_print_events")
def main():
  for i in range(20):
   print(i*2)

if __name__ == "__main__":
  main()
```

```python
print("03_wholesome_machine")

correct_affirmation = "I am capable of doing anything: I put my mind to."

def main():
    print("Welcome to the Wholesome Machine")
    while True:
      user_input = input("Please type the following affirmation:\n" + correct_affirmation + "\n> ")
      if user_input == correct_affirmation:
            print("That's right!")
            break
      else:
            print("Hmmm, that was not the affirmation. Please try again!")

if __name__ == "__main__":
    main()
```

```python
print("04_liftoff.")

def main():
  for i in range(10, 0, -1):
    print(i, end = "")
  print("Liftoff!")

if __name__ == "__main__":
  main()
```

```python
print("05_double_it")
def main():
  user_value = int(input("Please enter a number: "))
  while user_value < 100:
    user_value *= 2
  print(user_value * 2)
if __name__ == "__main__":
  main()
```
