```python
# Taking input from the user
name = input("Enter your name: ")
age = int(input("Enter your age (as an integer): "))
height = float(input("Enter your height (in float): "))

# Typecasting age to float and height to int
converted_age = float(age)
converted_height = int(height)

# Displaying the results
print("\n--- Converted Information ---")
print("Name:", name)
print("Age (as float):", converted_age)
print("Height (as int):", converted_height)
```

```python
import random

# Create a list of 5 random integers between 10 and 50
numbers = [random.randint(10, 50) for _ in range(5)]
print("Initial list:", numbers)

# Append a new number to the list
numbers.append(random.randint(10, 50))
print("After appending a new number:", numbers)

# Remove a number using remove()
# Let's remove the first element in the list by value
number_to_remove = numbers[0]
numbers.remove(number_to_remove)
print(f"After removing the number {number_to_remove}:", numbers)

# Delete an element using del
# Let's delete the element at index 1 (second element)
del numbers[1]
print("After deleting element at index 1:", numbers)

# Pop the last element
popped_element = numbers.pop()
print(f"After popping the last element ({popped_element}):", numbers)

# Slice and print the first 3 elements of the list
first_three = numbers[:3]
print("First 3 elements:", first_three)
```

```python
even_numbers = []

# Using for loop to generate even numbers between 1 and 50
for num in range(1, 51):
    if num % 2 == 0:
        even_numbers.append(num)

# Counting how many even numbers are in the list
count = len(even_numbers)

# Printing the list and the count
print("Even numbers between 1 and 50:", even_numbers)
print("Count of even numbers:", count)
```

```python
total_sum = 0

while True:
    num = float(input("Enter a positive number (0 or negative to stop): "))
    if num <= 0:
        break
    total_sum += num

print("Sum of all positive numbers entered:", total_sum)
```

```python
# Ask user to input a number
num = int(input("Enter a number: "))

# Check if the number is even or odd and apply the conditions
if num % 2 == 0 and num > 10:
    print("Even and greater than 10")
elif num % 2 == 0 and num <= 10:
    print("Even and 10 or less")
elif num % 2 != 0:
    print("Odd number")
```

```python
# Ask user to input a number
num = int(input("Enter a number: "))

# Define Boolean logic
is_even = num % 2 == 0
is_odd = not is_even
greater_than_10 = num > 10
less_or_equal_10 = num <= 10

# Use Boolean logic in conditions
if is_even and greater_than_10:
    print("Even and greater than 10")
elif is_even and less_or_equal_10:
    print("Even and 10 or less")
elif is_odd:
    print("Odd number")
```

```python
def generate_even_numbers(n):
    even_numbers = []
    for i in range(0, n):
        if i % 2 == 0:
            even_numbers.append(i)
    return even_numbers

# Call the function with n = 10 and print the result
result = generate_even_numbers(10)
print(result)
```

```python
# Ask the user to enter three numbers
num1 = float(input("Enter the first number: "))
num2 = float(input("Enter the second number: "))
num3 = float(input("Enter the third number: "))

# Use conditional statements to find the largest number
if num1 >= num2 and num1 >= num3:
    largest = num1
elif num2 >= num1 and num2 >= num3:
    largest = num2
else:
    largest = num3

# Print the largest number
print("The largest number is:", largest)
```
